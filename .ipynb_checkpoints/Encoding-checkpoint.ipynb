### Encoding of training dataset

```python
existing_df_3_copy = existing_df_3.copy()
```

```python
#label encoding

from sklearn.preprocessing import LabelEncoder

# Create a label encoder
label_encoder = LabelEncoder()

# Apply label encoding to each categorical column
for column in ['Continent', 'Industry', 'Country']:
    existing_df_3_copy[column] = label_encoder.fit_transform(existing_df_3_copy[column])

existing_df_3_copy.head()
```

```python
unique_values = existing_df_3_copy['Country'].unique()
unique_values

```

```python
#one hot encoding

columns_to_encode = ['Continent', 'Industry','Country']
df_encoded = pd.get_dummies(existing_df_3, columns=columns_to_encode, drop_first=True)


df_encoded
```

The 'Continent' variable has undergone numerical representation, exhibiting six distinct values ranging from 0 to 5. Similarly, the 'Industry' variable has been encoded numerically, featuring eleven unique values spanning from 0 to 10.


After label encoding 'Continent,' which resulted in 6 unique values, applying one-hot encoding with drop_first=True creates 5 binary columns. Similarly, for 'Industry,' with 11 unique values post label encoding, one-hot encoding results in the generation of 10 binary columns.


In ordinal encoding, assigning values to each unique category is a necessary step, adding a layer of complexity. Therefore, in situations where simplicity is preferred, label encoding proves to be a more straightforward choice.


Encoding columns with numerous unique values, such as "Country," can complicate the dataset by increasing dimensionality and potentially impacting machine learning model performance. The high cardinality may lead to sparse representations, hindering effective generalization.
