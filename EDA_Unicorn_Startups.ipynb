#### Exited Unicorn Companies

```python
# Assuming you want to plot a count of each country in a column called 'Top Countries'
top_countries = exited_df_3['Country'].value_counts().nlargest(10).index

# Filter the dataframe to include only the top 10 countires
df_top_countries = exited_df_3[exited_df_3['Country'].isin(top_countries)]

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")  # Set the style of the plot

ax = sns.countplot(data=df_top_countries, x='Country', color="darkseagreen", order=top_countries)

plt.title('Top 10 Countries Distribution of Exited Unicorn Companies', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Add counts to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', fontweight='bold', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

The majority of unicorn companies in the United States have exited, as indicated by the above chart. China comes in second.

```python
# Assuming you want to plot a count of each industry in a column called 'Top Industries'
top_industries = exited_df_3['Main Industry'].value_counts().nlargest(10).index

# Filter the dataframe to include only the top 10 industries
df_top_industries = exited_df_3[exited_df_3['Main Industry'].isin(top_industries)]

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")  # Set the style of the plot

ax = sns.countplot(data=df_top_industries, x='Main Industry', color="darkseagreen", order=top_industries)

plt.title('Top 10 Industries Distribution of Exited Unicorn Companies', fontsize=16, fontweight='bold')
plt.xlabel('Industry', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Add counts to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', fontweight='bold', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

#### Exited Unicorn Companies

```python
# Assuming you want to plot a count of each industry in a column called 'Top Industries'
top_industries = exited_df_3['Main Industry'].value_counts().nlargest(10).index

# Filter the dataframe to include only the top 10 industries
df_top_industries = exited_df_3[exited_df_3['Main Industry'].isin(top_industries)]

# Create a bar chart with hue='Main Reason'
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")  # Set the style of the plot

ax = sns.countplot(data=df_top_industries, x='Main Industry', hue='Main Reason', palette='turbo_r', order=top_industries)

plt.title('Main Exit Reasons for the Top 10 Industries', fontsize=16, fontweight='bold')
plt.xlabel('Industry', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Remove counts from the bars
for p in ax.patches:
    ax.annotate('', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', fontweight='bold', xytext=(0, 5),
                textcoords='offset points')

plt.legend(title='Main Reason', title_fontsize='12', loc='upper right', bbox_to_anchor=(1.15, 1))

plt.show()
```

According to the above chart, bankruptcy is the reason for the company's exit in all of the top ten industries of the exited unicorn companies. The second most important reason for an IPO. Only unicorn companies in the software and technology industries, as well as the e-commerce and retail industries, have exited due to devaluation. 

```python
# Create a stacked bar chart
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")  # Set the style of the plot

# Assuming you want to plot a count of each company in a column called 'Exit_Reason' stacked by 'Year'
ax = sns.countplot(data=exited_df_3, x='Year', hue='Main Reason', palette='turbo_r', dodge=True)

plt.title('Yearly Distribution of Exited Unicorn Companies by Exit Reason', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')

plt.legend(title='Exit Reason', title_fontsize='12', fontsize='10', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

The main reason for exiting unicorns, according to the chart from 2017 to 2023, is bankruptcy. Except for devaluation, the four main reasons to exit are higher in 2021 than in previous years. Only in 2020 will there be a devaluation.
