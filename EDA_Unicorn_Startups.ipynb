```python
# Assuming you want to plot a count of each country in a column called 'Top Countries'
top_countries = exited_df_3['Country'].value_counts().nlargest(10).index

# Filter the dataframe to include only the top 10 countires
df_top_countries = exited_df_3[exited_df_3['Country'].isin(top_countries)]

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")  # Set the style of the plot

ax = sns.countplot(data=df_top_countries, x='Country', color="darkseagreen", order=top_countries)

plt.title('Top 10 Countries Distribution of Exited Unicorn Companies', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Add counts to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', fontweight='bold', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

#### Emerging Unicorn Companies

```python
# Assuming 'Emerging_Companies.csv' has columns 'Lead Investors Include' and 'Country'
emerging_df_3 = pd.read_csv("Emerging_Companies.csv", encoding='unicode_escape')

# Splitting the 'InvestorColumn' and creating new rows
investor_list = emerging_df_3['Lead Investors Include'].str.split(', ')
investor_df = pd.DataFrame({
    'Investor': investor_list.sum(),  # Flatten the list of lists
    'Country': emerging_df_3['Country'].repeat(investor_list.apply(len))
})

# Counting the number of investors for each country
investor_counts = investor_df.groupby('Country').count().reset_index()

# Renaming the columns for clarity
investor_counts.columns = ['Country', 'Investor_Count']

# Display the result
print(investor_counts)
```

```python
# Selecting the top 10 countries
top_10_countries = investor_counts.sort_values(by='Investor_Count', ascending=False).head(10)

# Creating a bar chart with count values on top
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
bars = plt.bar(top_10_countries['Country'], top_10_countries['Investor_Count'], color='lightsalmon')
plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Number of Investors', fontsize=12, fontweight='bold')
plt.title('Top 10 Countries by Number of Investors in Emerging Unicorn Companies', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Adding count values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, round(yval, 1), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()

# Display the bar chart
plt.show()
```

### Multivariate Analysis:

Connection between number of investors in a country and the number of startups

#### Existing Unicorn Companies

```python
# Assuming 'Emerging_Companies.csv' has columns 'Lead Investors Include' and 'Country'
existing_df_3 = pd.read_csv("Unicorn_Companies.csv", encoding='unicode_escape')

# Splitting the 'Lead Investors Include' column and creating new rows
investor_list = existing_df_3['Lead Investors Include'].str.split(', ')
investor_df = pd.DataFrame({
    'Investor': investor_list.sum(),  # Flatten the list of lists
    'Country': existing_df_3['Country'].repeat(investor_list.apply(len))
})

# Counting the number of investors for each country
investor_counts = investor_df.groupby('Country').count().reset_index()
investor_counts.columns = ['Country', 'Investor_Count']

# Counting the number of companies for each country
company_counts = existing_df_3['Country'].value_counts().reset_index()
company_counts.columns = ['Country', 'Company_Count']

# Merging the two dataframes on the 'Country' column
result_df = pd.merge(investor_counts, company_counts, on='Country', how='outer').fillna(0)

# Display the result
print(result_df)
```

```python
df = pd.DataFrame(result_df)

# Sorting by Investor_Count for better visualization
df = df.sort_values('Investor_Count', ascending=False)

# Selecting the top 10 countries
df_top10 = df.head(10)

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))

color = 'lightsalmon'
ax1.set_xlabel('Country', fontsize=12, fontweight='bold')
ax1.set_ylabel('Investor Count', color=color, fontsize=12, fontweight='bold')
ax1.plot(df_top10['Country'], df_top10['Investor_Count'], color=color, marker='o', label='Investor Count')
ax1.tick_params(axis='y', labelcolor=color)
plt.xticks(rotation=45, ha='right')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'lightslategrey'
ax2.set_ylabel('Company Count', color=color, fontsize=12, fontweight='bold')  # we already handled the x-label with ax1
ax2.bar(df_top10['Country'], df_top10['Company_Count'], color=color, label='Company Count', alpha = 0.7)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Top 10 Investor and Company Count by Country', fontsize=16, fontweight='bold')
plt.show()
```

The graph above demonstrates that as the number of companies in every country declines, so does the number of investors.
